```python
# Create a list of 10 integers
numbers = [5, 12, 3, 8, 20, 7, 1, 15, 9, 4]

print("Original list:", numbers)

# Add an element
numbers.append(25)
print("After adding element (25):", numbers)

# Remove an element
numbers.remove(3)
print("After removing element (3):", numbers)

# Sort in ascending order
ascending = sorted(numbers)
print("Ascending order:", ascending)

# Sort in descending order
descending = sorted(numbers, reverse=True)
print("Descending order:", descending)

# Find sum of elements
total_sum = sum(numbers)
print("Sum of elements:", total_sum)

# Subtract all elements one by one
result = numbers[0]
for num in numbers[1:]:
    result -= num
print("Result after subtraction:", result)

# Find maximum and minimum
maximum = max(numbers)
minimum = min(numbers)

print("Maximum value:", maximum)
print("Minimum value:", minimum)
```

```python
# Create a dictionary
student = {
    "name": "Ali",
    "age": 20,
    "grade": "A"
}

print("Original dictionary:", student)

# Update value
student["age"] = 21
print("After updating age:", student)

# Add new key-value
student["city"] = "Karachi"
print("After adding city:", student)

# Remove a key-value
del student["grade"]
print("After deleting grade:", student)

# Another way to remove
student.pop("city")
print("After removing city:", student)

# Check if key is present
if "name" in student:
    print("Key 'name' is present")
else:
    print("Key 'name' is not present")

# Check another key
if "grade" in student:
    print("Key 'grade' is present")
else:
    print("Key 'grade' is not present")
```
