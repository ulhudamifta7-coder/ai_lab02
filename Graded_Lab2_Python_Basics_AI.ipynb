```python
# Task 1.7: Display all attributes
class Student:
    def __init__(self, student_id, student_name, student_class):
        self.student_id = student_id
        self.student_name = student_name
        self.student_class = student_class

    def display(self):
        for key, value in self.__dict__.items():
            print(key, ":", value)

s = Student(1, "Ali", "BSCS")
s.display()
```

```python
# Task 1.8: Two instances
class Student:
    def __init__(self, student_id, student_name, student_class):
        self.student_id = student_id
        self.student_name = student_name
        self.student_class = student_class

    def display(self):
        print("ID:", self.student_id)
        print("Name:", self.student_name)
        print("Class:", self.student_class)
        print("-------------------")

student1 = Student(1, "Ali", "BSCS")
student2 = Student(2, "Sara", "BBA")

student1.display()
student2.display()
```

```python
# Task 1.9: Memory allocation
import sys

class Student:
    def __init__(self, name):
        self.name = name

s = Student("Ali")

print("Memory size of object:", sys.getsizeof(s))
print("Memory size of attributes:", sys.getsizeof(s.__dict__))
```

```python
# Task 1.10: Memory optimization using __slots__
import sys

class Student:
    __slots__ = ['name']   # reduces memory

    def __init__(self, name):
        self.name = name

s = Student("Ali")

print("Memory after using __slots__:", sys.getsizeof(s))
```

```python
# ============================================
# LAB TASK 2: Object-Oriented Design & Area Comparison
# ============================================

# 1. Define Classes

class Trapezoid:
    def __init__(self, a, b, height):
        self.a = a              # parallel side 1
        self.b = b              # parallel side 2
        self.height = height    # height

    def area(self):
        return 0.5 * (self.a + self.b) * self.height


class Parallelogram:
    def __init__(self, base, height):
        self.base = base        # base
        self.height = height    # height

    def area(self):
        return self.base * self.height


# 2,3,4. Comparison + Dictionary + Multiple Inputs

# Dictionary to store areas
areas = {}

# Multiple sets of dimensions
shapes = [
    ("Trapezoid_1", Trapezoid(5, 7, 4)),
    ("Parallelogram_1", Parallelogram(6, 4)),
    ("Trapezoid_2", Trapezoid(8, 10, 5)),
    ("Parallelogram_2", Parallelogram(7, 3))
]

# Calculate and store areas
for name, shape in shapes:
    areas[name] = shape.area()


# Display all areas
print("Areas of Shapes:")
for shape_name, area_value in areas.items():
    print(shape_name, "=", area_value)


# 5. Find Shape with Maximum Area

max_shape = max(areas, key=areas.get)

print("\nShape with Maximum Area:")
print(max_shape, "=", areas[max_shape])
```
